```python
import numpy as np 
import pandas as pd 
import sklearn
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data.csv')
df
```

```python
x = df.iloc[:, :-1].values 
x
```

```python
y = df.iloc[:, 3].values 
y
```

Pandas-Handling missing values


```python
df['Age'] = df['Age'].fillna(value = df['Age'].mean())
df['Salary'] = df['Salary'].fillna(value = df['Salary'].mean())
df
```

 SKlearn -Handling missing values


```python
from sklearn.impute import SimpleImputer
im = SimpleImputer(missing_values = np.nan, strategy='median')
im.fit(x[:,1:3])
x[:,1:3] = im.transform(x[:,1:3])
x
```
