# ***matplotlib***






```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/prac1/Toyota.csv')
df
```

```python
km1=df.KM
age1=df.Age
m1=df.Price
```

```python
plt.bar(age1,km1,color='grey') 
plt.legend()
plt.ylabel('km')
plt.xlabel('age')

```

```python
plt.hist([age1,km1],bins=100 ,rwidth=0.8)
```

```python
plt.scatter(km1,m1,label='age',color='b')
plt.ylabel('price')
plt.xlabel('km')


```

```python
plt.pie(age1,m1)
```

```python
plt.plot(age1)

plt.plot(m1)
plt.plot(km1)


```
